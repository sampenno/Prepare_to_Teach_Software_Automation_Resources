__4. In a markdown block, outline the Machine Learning Linear Regression algorithm.__

The Machine Learning Linear Regression Algorithm takes in a linear data set. It has a cost function that is able to measure the loss of each data point to create a cost for the whole model based on $y = mx + b$ . It then applies a gradient descent algorithm that loops through all the possible coefficients $m$ and y intercepts $b$ to find the lowest possible cost for the model. The algorithm returns the lowest $b$ and $m$, which can be used for predictions by applying $y = mx + b$.

__7. Refering to the below model, describe underfitting in Linear Regression.__

````python
x_train = np.array([1, 1.5, 1.8, 2, 2.2])
y_train = np.array([[200], [250], [260], [280], [500]])


m = 200
b = 10

tmp_f_mb = compute_model_output(x_train, m, b,)

plt.plot(x_train, tmp_f_mb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.show()
````

Underfitting is when noise or outliers negatively affect the model performance and intuition. The data point (500,2.2) is a clear outlier as it is in the top quantile. The function of this model is $y = mx + b$; when the model has been 'fitted' to the data set, this data point has caused the coefficient of the model ($m$) to be too high, resulting in poor intuition to most of the training data and new or testing data. By removing outliers in the top quantile, including data point (500,2.2), when the model is re-fitted, the y-intercept ($b$) would be slightly different, and the coefficient would be significantly lower. This would improve the model's intuition on the training data, testing data and new data.

Because of the high variance gradient, descent could settle on the local lowest cost, not the global lowest cost.
